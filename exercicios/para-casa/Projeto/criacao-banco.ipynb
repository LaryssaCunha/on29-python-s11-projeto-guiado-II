```python
import csv
import sqlite3
```

```python
file = open("brasileirao_serieA_finalCampeonato.csv")
```

```python
conteudo = csv.reader(file)
```

```python
connection = sqlite3.connect("brasileirao2013_2023.db")
cursor = connection.cursor()
```

```python
cursor.execute('''CREATE TABLE brasileirao2013_2023 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    ano INT,
    posicao INT,
    time TEXT,
    UF TEXT,
    pontos INT,
    partidas_jogadas INT,
    vitorias INT,
    empates INT,
    derrotas INT,
    gols_pro INT,
    gols_contra INT,
    saldo_gols INT,
    cartao_amarelo INT,
    cartao_vermelho INT,
    aproveitamento FLOAT,
    ultimas_jogadas INT
)''')
```

```python
inserir_conteudo = 'INSERT INTO brasileirao2013_2023(ano,posicao,time,UF,pontos,partidas_jogadas,vitorias,empates,derrotas,gols_pro,gols_contra,saldo_gols,cartao_amarelo,cartao_vermelho,aproveitamento,ultimas_jogadas) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)'
```

```python
cursor.executemany(inserir_conteudo, conteudo)
```

```python
selecionar_tudo = "SELECT * FROM brasileirao2013_2023"
```

```python
entradas = cursor.execute(selecionar_tudo)
```

```python
for entrada in entradas:
    print(entrada)
```

```python
connection.commit()
cursor.close()
connection.close()
```
